Raman spectroscopy can, in principle, be used to estimate the diameter of CNTs. This is because CNTs have a radial breathing mode (RBM) whose frequency depends on their diameter. In particular, for a single-walled tube (it differs slightly for a multi-walled):
$$ω_{RBM} \approx \frac{248}{d}$$ 
where $ω_{RBM}$ is the frequency of the RBM peak in cm^-1 and $d$ is the diameter in nm.

Below: Raman spectrum for a typical SWCNT

![](../images/150716_CNT_Raman.jpg)


However, to actually observe this peak requires (see Yao 2009):
1. A substrate that does not produce a large background spectrum. Quartz does not work well for this purpose (see fig below). Thus, the tubes must either be transferred to a different substrate or grown on top of another material (e.g. thin layer of silica). 
2. The RBM signal is only detectable in tubes whose electronic transition energy matches the energy of the laser. For example, Yao found that only 6% of their tubes were "active".

To test the technique, I took 20 spectra at random locations on a standard growth (S8, density ~1/micron). All of these spectra except for one showed only the background signal due to quartz.

```python
shift, count = loadtxt('../data/150715_Raman_S8_no_CNT.txt', unpack = True)
plot(shift, count)
xlim(100,3000)
xlabel('Raman shift (cm^-1)');
```

The one outlier was the following:

```python
shift, count = loadtxt('../data/150715_Raman_S8_with_CNT.txt', unpack = True)

_, (ax1, ax2) = subplots(1, 2, figsize = (14,5))

ax1.plot(shift, count)
ax1.set_xlim(100,3000)
ax1.set_xlabel('Raman shift (cm^-1)')

ax2.plot(shift, count)
ax2.set_xlim(100,3000)
ax2.set_ylim(0,2000)
ax2.set_xlabel('Raman shift (cm^-1)');
```

The additional peaks, especially those around 1600 and 2700, may be evidence for a nanotube. But there is no clear detection of the RBM, either because it is hidden beneath the background signal or it was never excited in the first place. In any case, it seems that Ramans is practically useless for our purposes.
