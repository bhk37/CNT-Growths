This is a preliminary characterization of the CNT furnace temperature profile using an IR gun.

## Set-up
- I mounted the IR gun on a tripod and aimed it at a glass cap in the center of the furnace. The purpose of the cap is to simulate the thermal properties of the quartz chips, while providing a larger cross-section at which to focus the gun. 
- I aligned the gun using a half tube (2 ft). This is because the lasers on the gun would hit the sides of a full tube.
- It helped to align the lasers first on the ends of the tube (using plastic end caps), then verify that they were centered on the cap in the middle of the tube.


## Measurement 1 
- 100 degree intervals from 100 degrees to 1000 degrees
- I placed the cap in the exact center of the furnace (marked by the notches in the styrofoam). 
- I adjusted the temperature of the furnace manually, i.e. restarting the program for each temperature. 
- I set the gun to measure continuously (mode -> hold on) and recorded the temperature after about 10 minutes of settling time.

```python
set_temp, readout = loadtxt('../data/150527_furnace_characterization_1.txt', unpack = True)

plot(linspace(0,1100), linspace(0,1100), 'k--', label = 'Theory')
plot(set_temp, readout, 's', label = 'Experiment')
xlabel('Set Temp (C)')
ylabel('Readout Temp (C)')
legend(loc = 'best')
xlim(0,1050);
```
