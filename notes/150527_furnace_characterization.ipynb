This is a preliminary characterization of the CNT furnace temperature profile using an IR gun.

## Set-up
The IR gun is mounted on a tripod outside of the furnace and aimed down the center of the furnace at a glass cap within the tube. The purpose of the cap is to simulate the thermal properties of the quartz chips, while providing a larger cross-section at which to focus the gun. 

## Measurement 1 - 100 degree steps
The cap is in the exact center of the furnace (marked by the notches in the styrofoam). I set the gun to record continuously (mode -> hold on) and adjusted the temperature of the furnace manually. For each set point, I recorded the temperature after after around 10 minutes, which is about the time it took to settle.

```python
set_temp, readout = loadtxt('../data/150527_furnace_characterization_1.txt', unpack = True)

figure(figsize = (7,5))
plot(linspace(0,1100), linspace(0,1100), 'k--', label = 'Theory')
plot(set_temp, readout, 's', label = 'Experiment')
xlabel('Set Temp (C)', fontsize = 14)
ylabel('Readout Temp (C)', fontsize = 14)
legend(loc = 'best')
xlim(0,1050);
```
